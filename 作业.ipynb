# 2. 统计全班的成绩

```python
from pandas import DataFrame,Series
data = {'Chinese': [66, 95, 93, 90,80], 'Math': [30, 98, 96, 77, 90], 'English': [65, 85, 92, 88, 90]}
df2 = DataFrame(data, index=['ZhangFei', 'GuanYu', 'LiuBei', 'DianWei', 'XuChu'], columns=['Chinese', 'Math', 'English'])
df2.describe()
# print(df2)
```

```python
# 求各人的成绩和
print(df2)
df2['Sum'] = df2.apply(lambda x: x.sum(), axis=1)
```

```python
print(df2)
df2 = df2.sort_values('Sum', ascending=False)
```

```python
df2.reset_index(inplace=True)
df2= df2.rename(columns={'index':'Name'})
print(df2)
```

# 3. 对汽车质量数据进行统计

```python
df3 = pd.read_csv('car_complain.csv')
df3.head()
```

```python
def f(x):
    x = x.replace('一汽-大众','一汽大众')
    return x
```

```python
# pd.get_dummies(df3['brand'])
df3['brand'] = df3['brand'].apply(f)
# df3.loc[df3['brand'] == '一汽-大众','brand'] = '一汽大众'
df3 = df3.drop('problem',axis=1).join(df3.problem.str.get_dummies(','))
df3.head()
```

```python
result = df3.groupby(['brand'])['id'].agg(['count'])

result
```

```python
tages = df3.columns[7:]
# result5 = df3.groupby(['brand'])['A11','A114'].agg(['count'])
# result5
result2 = df3.groupby(['brand'])[tages].agg(['sum'])
result2 
```

```python
result2 = result.merge(result2,left_index = True,right_index=True, how='left')
result2
```

```python
result2.to_csv('result.csv')
```

```python
#按照count来排序
result2.sort_values('count',ascending=False)
```
